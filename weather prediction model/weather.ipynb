```python
import requests
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
API_KEY = "YOUR_API_KEY"  # Replace with your OpenWeatherMap API key
CITY = "Delhi"
DAYS = 7
```

```python
# ---- FETCH WEATHER DATA ----
def fetch_weather_data(city):
    print(f"Fetching weather data for {city}...")

    # OpenWeatherMap 5-day/3-hour forecast API
    url = f"http://api.openweathermap.org/data/2.5/forecast?q={city}&appid={API_KEY}&units=metric"

    response = requests.get(url)
    if response.status_code != 200:
        print("Error fetching data:", response.status_code)
        return None
    
    data = response.json()
    return data
```

```python
# ---- PROCESS DATA ----
def process_weather_data(data):
    dates = []
    temps = []

    for entry in data["list"]:
        # Convert timestamp to readable date
        dt = datetime.fromtimestamp(entry["dt"])
        temp = entry["main"]["temp"]

        # Collect one entry per day (midday)
        if dt.hour == 12:
            dates.append(dt.strftime("%d %b"))
            temps.append(temp)

    return dates, temps
```

```python
# ---- VISUALIZE ----
def plot_temperature_trends(dates, temps, city):
    plt.figure(figsize=(10, 5))
    plt.plot(dates, temps, marker='o', linestyle='-', linewidth=2)
    plt.title(f"Daily Temperature Trends in {city}")
    plt.xlabel("Date")
    plt.ylabel("Temperature (Â°C)")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# ---- MAIN ----
def main():
    data = fetch_weather_data(CITY)
    if data:
        dates, temps = process_weather_data(data)
        plot_temperature_trends(dates, temps, CITY)
    else:
        print("No data available.")
```

```python
if __name__ == "__main__":
    main()
```
